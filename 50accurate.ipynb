```python
def mean_average_precision(recommended, relevant):
    average_precisions = []
    for rec, rel in zip(recommended, relevant):
        if not rel:
            continue
        precisions = []
        num_hits = 0
        for i, item in enumerate(rec):
            if item in rel:
                num_hits += 1
                precisions.append(num_hits / (i + 1))
        average_precisions.append(sum(precisions) / len(rel))
    return sum(average_precisions) / len(average_precisions)

# Example usage
recommended = [[1, 2, 3], [4, 5, 6]]
relevant = [[2, 3], [5, 7]]
print("MAP:", mean_average_precision(recommended, relevant))
```

```python
import numpy as np

def ndcg(recommended, relevant, k=10):
    def dcg(recommended, relevant, k):
        recommended = recommended[:k]
        gain = [1 if item in relevant else 0 for item in recommended]
        discounts = [np.log2(i + 2) for i in range(len(gain))]
        return sum([g / d for g, d in zip(gain, discounts)])

    scores = []
    for rec, rel in zip(recommended, relevant):
        idcg = dcg(rel, rel, k)  # Ideal DCG
        actual_dcg = dcg(rec, rel, k)
        scores.append(actual_dcg / idcg if idcg > 0 else 0)
    return np.mean(scores)

# Example usage
print("NDCG:", ndcg(recommended, relevant, k=3))
```

```python
def mean_reciprocal_rank(recommended, relevant):
    reciprocal_ranks = []
    for rec, rel in zip(recommended, relevant):
        for rank, item in enumerate(rec):
            if item in rel:
                reciprocal_ranks.append(1 / (rank + 1))
                break
        else:
            reciprocal_ranks.append(0)
    return sum(reciprocal_ranks) / len(reciprocal_ranks)

# Example usage
print("MRR:", mean_reciprocal_rank(recommended, relevant))
```
